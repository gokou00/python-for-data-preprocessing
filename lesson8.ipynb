# Handling Mixed and DateTime Variables 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Creating our dummy dataset

name = ["Jon","Nick","Ben","Sally","Alice","Josh"]
education = [9,"Graduate",7,"Graduate","PhD",8]
std = {"name":name,"Qualification":education}

student_df = pd.DataFrame(std)
student_df.head()
```

```python
#Creating a numeric column for only the numeric values in the Qualification column

student_df["q_numeric"] = pd.to_numeric(student_df["Qualification"],errors="coerce",downcast="integer")
```

```python
#Creating a categorical column for only the string values in the Qualification column

student_df["q_categoric"] = np.where(student_df["q_numeric"].isnull(),student_df["Qualification"],np.nan)

student_df.head()
```

```python
#Downloading the titanic dataset

titanic_data = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

titanic_data.head()
```

```python
#Filtering for only the ticket and cabin columns

titanic_data = titanic_data[["Ticket","Cabin"]]

titanic_data.head()
```

```python
#Creating 2 different columns for the ticket column one for number and the other for strings

titanic_data["Ticket_Num"] = titanic_data["Ticket"].str.extract("(\d+)")
titanic_data["Ticket_Cat"] = titanic_data["Ticket"].str[0]

titanic_data[["Ticket","Ticket_Num","Ticket_Cat"]].head()
```

# Handling Date Data Type

```python
#Downloading the dataset of tesla stock price

tesla_stock = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/tesla-stock-price.csv")
tesla_stock = tesla_stock.shift(-1)
tesla_stock.dropna(inplace=True)
tesla_stock.head()
```

```python
#Using the to_datetime function to convert the date column which is now a string to a date column

tesla_stock["date"] = pd.to_datetime(tesla_stock["date"])
```

```python
# Creating a new column week from the newly converted date column

tesla_stock["week"] = tesla_stock["date"].dt.week

tesla_stock[["date","week"]].head()
```

```python
#Creating the month number from the date column

tesla_stock["month"] = tesla_stock["date"].dt.month
tesla_stock[["date","month"]].head()
```

```python
#Finding the day of the month as an integer using dt.day

tesla_stock["day_month"] = tesla_stock["date"].dt.day

tesla_stock[["date","day_month"]].head()
```

```python
#Getting the name of the day of the week 

tesla_stock["day_week"] = tesla_stock["date"].dt.weekday

tesla_stock[["date","day_week"]].head()
```

```python
#Finding the difference between two dates

diff = tesla_stock["date"].iloc[0] - tesla_stock["date"].iloc[4]

print(tesla_stock["date"].iloc[0])
print(tesla_stock["date"].iloc[4])
print(diff)
```

# Handling Time Data Type

```python
#Downloading the bike dataset

bike_data = pd.read_csv("https://raw.githubusercontent.com/QROWD/TR/master/datasets/bike.csv")

bike_data.dropna(inplace=True)
bike_data.head()
```

```python
#Converting the timestamp to a datetime column

bike_data["timestamp"] = pd.to_datetime(bike_data["timestamp"])
```

```python
#Creating an hour, minute and second columns from the timestamp column


bike_data["hour"] = bike_data['timestamp'].dt.hour
bike_data["minute"] = bike_data["timestamp"].dt.minute
bike_data["second"] = bike_data["timestamp"].dt.second


bike_data.shift(-50).head(20)
```

```python
#Getting the time part from the timestamp

bike_data["time"] = bike_data["timestamp"].dt.time

bike_data.shift(-50).head(20)
```
